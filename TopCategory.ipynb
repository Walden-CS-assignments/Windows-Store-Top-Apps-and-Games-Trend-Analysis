# 合并相似种类

```python
df.loc[df['Category'].str.contains('Business')] = 'Business'
df.loc[df['Category'].str.contains('Developer tools')] = 'Developer tools'
df.loc[df['Category'].str.contains('Personalization')] = 'Personalization'
df.loc[df['Category'].str.contains('Education')] = 'Education'
df.loc[df['Category'].str.contains('Utilities & tools')] = 'Utilities & tools'
df.loc[df['Category'].str.contains('Multimedia design')] = 'Multimedia design'
df.loc[df['Category'].str.contains('Books & reference')] = 'Books & reference'
df.loc[df['Category'].str.contains('Security')] = 'Security'
df.loc[df['Category'].str.contains('Kids & family')] = 'Kids & family'
df.loc[df['Category'].str.contains('Lifestyle')] = 'Lifestyle'
df.loc[df['Category'].str.contains('News & weather')] = 'News & weather'
df.loc[df['Category'].str.contains('Personal finance')] = 'Personal finance'
# .replace(df['Category'].str.contains('Business') , 'Business')
```

```python
df['Category'].value_counts()
```

```python
Top_Cat = df['Category'].value_counts().head(20).index
Top_Cat = Top_Cat.tolist()
```

```python
Top_Cat
```

```python
Top_CatNum = df['Category'].value_counts().head(20)
Top_CatNum = Top_CatNum.tolist()
```

```python
Top_CatNum
```

```python
Top_CatNum.reverse()
Top_Cat.reverse()

plt.barh(Top_Cat, Top_CatNum)

plt.title("Most Popular Category")

plt.xlabel("Number of Apps for each Category")

plt.tight_layout()

plt.savefig('Top_20_Category.png')

plt.show()
```
